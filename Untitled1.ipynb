```python
import pandas as pd
df = pd.read_csv('/content/pedidos.csv')
df.head(5)
```

```python
pedidos=df.iloc[:,0:6]
pedidos.head(5)
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Comunas_de_Chile'
tables = pd.read_html(url)
```

```python
tables[0]
```

```python
comunas = tables[0]
```

```python
comunas.info()
```

```python
columnas = ['CUT (Cdigo nico Territorial)', 'Nombre', 'Provincia', 'Regin']
comunas = comunas[columnas]
```

```python
comunas.head()
```

```python
comunas = comunas.rename(columns={'CUT (Cdigo nico Territorial)': 'COMUNA'})
```

```python
comunas.head()
```

```python
pedidos.head()
```

```python
despachos.head()
```

```python
pedidos
```

```python
ventas = pedidos.groupby('Correo')['Monto total'].sum()
```

```python
ventas.head(10)
```

```python
cliente_top = ventas.sort_values(ascending=False).head(1)
```

```python
ventas.to_excel("informe_ventas.xlsx")
```

```python
ventas.to_csv("informe_ventas.csv")
```

```python
print(cliente_top)
```

```python
resumen = despachos.merge(pedidos, on='PEDIDO', how='inner')
```

```python
resumen.to_csv('resumen_ventas.csv')
```

```python
resumen.info()
```

```python
resumen = resumen.merge(comunas, on='COMUNA', how='inner')
```

```python
ventas_region = resumen.groupby('Regin')['Monto total'].sum()
```

```python
ventas_region.to_csv('reporte_region.csv')
```

```python
region_top = ventas_region.sort_values(ascending=False)
```

```python
region_top.to_csv('reporte_xregion.csv')
```
